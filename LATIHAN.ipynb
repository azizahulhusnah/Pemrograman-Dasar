Perintah ziza_df.tail() dalam Python digunakan untuk menampilkan 5 baris terakhir dari DataFrame ziza_df secara default. Fungsi .tail() sangat berguna untuk memeriksa data di bagian akhir dataset, misalnya untuk melihat apakah ada nilai yang hilang, duplikat, atau pola tertentu yang muncul di baris terakhir. Sama seperti .head(), penggunaan .tail() hanya menampilkan data dan tidak mengubah isi DataFrame. Jika ingin menampilkan jumlah baris yang berbeda, misalnya 10 atau 20 baris terakhir, cukup menambahkan angka sebagai argumen, misalnya ziza_df.tail(10).
